```python
import math
import random


def QuickSort(A):
    if len(A) <= 1:
        return A
    L=[]
    R=[]
    p=random.choice(range(len(A)))
    E=[A[p]]
    for i in range(len(A)):
        if i==p :
            continue
        if A[i]<A[p]:
            L.append(A[i])
        else:
            R.append(A[i])
    return QuickSort(L)+E+QuickSort(R)

# numlist=[5,4,2,0,3,8,5,1,7]
numlist=[]

for i in range(30):
    u=int(random.uniform(0,100))    
    numlist.insert(0,u)

B=QuickSort(numlist)
B
```

```python
def bucket_sort(A, min_value, max_value):
    buckets = [[] for _ in range(min_value, max_value+1)]
    for x in A:
        buckets[x].append(x)
    sorted_arr = []
    for bucket in buckets:
        for item in bucket:
            sorted_arr.append(item)
    return sorted_arr
arr = [5,1,2,7,3,9,4,0,6,8]
sorted_arr = bucket_sort(arr, 0, 9)
print(sorted_arr)
```

```python
def getDigits(x, base):
    digits = []
    while x > 0:
        digits.append( x % base )
        x = (x / base).__trunc__()
    return digits
        

class myInt:
    def __init__(self,x, base=10, keyDigit=0):
        self.digits = getDigits(x, base)
        self.keyDigit = keyDigit
        self.value = x
        
    def key(self):
        if len(self.digits) > self.keyDigit:
            return self.digits[self.keyDigit]
        return 0
    
    def updateKeyDigit(self,p):
        self.keyDigit = p
        
    def getValue(self):
        return self.value

def bucketSortBase(A, base):
    buckets = [[] for _ in range(base)]
    for x in A:
        buckets[x.key()].append(x)# implement it by yourself (根据指定位置放进对应的桶里)
    sorted_arr = []
    for bucket in buckets: # From this
        for item in bucket: #
            sorted_arr.append(item)# implement it by yourself (按照顺序合并所有的桶)
    return sorted_arr

def radix_sort(A, n_digits, base):
    B = [ myInt(x, base=base) for x in A ]
    for j in range(n_digits):
        for x in B:
            x.updateKeyDigit(j) # implement it by yourself (指定比较的位置)
        B = bucketSortBase(B,base)    # implement it by yourself (桶排序)
    B = [x.getValue() for x in B]
    return B


arr = [523,123,4,33,12,532,642,763]
sorted_arr = radix_sort(arr, 3, 10)
print(sorted_arr)
```
